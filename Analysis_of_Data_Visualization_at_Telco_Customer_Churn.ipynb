# **Data Description**

```
ABOUT DATA
----------
A fictional telco company that provided home phone and Internet services to 7043 customers in California in Q3.

COLUMNS EXPLANATION
-------------------
CustomerID: A unique ID that identifies each customer.

Count: A value used in reporting/dashboarding to sum up the number of customers in a filtered set.

Country: The country of the customer’s primary residence.

State: The state of the customer’s primary residence.

City: The city of the customer’s primary residence.

Zip Code: The zip code of the customer’s primary residence.

Lat Long: The combined latitude and longitude of the customer’s primary residence.

Latitude: The latitude of the customer’s primary residence.

Longitude: The longitude of the customer’s primary residence.

Gender: The customer’s gender: Male, Female

Senior Citizen: Indicates if the customer is 65 or older: Yes, No

Partner: Indicate if the customer has a partner: Yes, No

Dependents: Indicates if the customer lives with any dependents: Yes, No. Dependents could be children, parents, grandparents, etc.

Tenure Months: Indicates the total amount of months that the customer has been with the company by the end of the quarter specified above.

Phone Service: Indicates if the customer subscribes to home phone service with the company: Yes, No

Multiple Lines: Indicates if the customer subscribes to multiple telephone lines with the company: Yes, No

Internet Service: Indicates if the customer subscribes to Internet service with the company: No, DSL, Fiber Optic, Cable.

Online Security: Indicates if the customer subscribes to an additional online security service provided by the company: Yes, No

Online Backup: Indicates if the customer subscribes to an additional online backup service provided by the company: Yes, No

Device Protection: Indicates if the customer subscribes to an additional device protection plan for their Internet equipment provided by the company: Yes, No

Tech Support: Indicates if the customer subscribes to an additional technical support plan from the company with reduced wait times: Yes, No

Streaming TV: Indicates if the customer uses their Internet service to stream television programing from a third party provider: Yes, No. The company does not charge an additional fee for this service.

Streaming Movies: Indicates if the customer uses their Internet service to stream movies from a third party provider: Yes, No. The company does not charge an additional fee for this service.

Contract: Indicates the customer’s current contract type: Month-to-Month, One Year, Two Year.

Paperless Billing: Indicates if the customer has chosen paperless billing: Yes, No

Payment Method: Indicates how the customer pays their bill: Bank Withdrawal, Credit Card, Mailed Check

Monthly Charge: Indicates the customer’s current total monthly charge for all their services from the company.

Total Charges: Indicates the customer’s total charges, calculated to the end of the quarter specified above.

Churn Label: Yes = the customer left the company this quarter. No = the customer remained with the company. Directly related to Churn Value.

Churn Value: 1 = the customer left the company this quarter. 0 = the customer remained with the company. Directly related to Churn Label.

Churn Score: A value from 0-100 that is calculated using the predictive tool IBM SPSS Modeler. The model incorporates multiple factors known to cause churn. The higher the score, the more likely the customer will churn.

CLTV: Customer Lifetime Value. A predicted CLTV is calculated using corporate formulas and existing data. The higher the value, the more valuable the customer. High value customers should be monitored for churn.

Churn Reason: A customer’s specific reason for leaving the company. Directly related to Churn Category.
```

### **1.   Visualiasi untuk menemukan perbedaan persentase dari `customer yang churn` dan `customer yang tidak churn`**

````python
#Pie chart
#melakukan grouping
data = df.groupby('Churn Label', as_index=False).count()
data
````

````python
df['Churn Label'].value_counts()
````

````python
fig = plt.subplots(figsize=(10,6))

plt.pie(data['Count'], labels=data['Churn Label'], autopct='%1.2f%%', pctdistance=0.2, wedgeprops=dict(width=0.2))
plt.xlabel('Customer Churn')
plt.ylabel('Percentage')
plt.title('Percentage of Customer Churn')

#legend
label_mapping = {'Yes': 'Customer yang Churn', 'No': 'Customer yang tidak Churn'}
legend_labels = [label_mapping[label] for label in data['Churn Label']]
plt.legend(legend_labels)

plt.show()
````

#### **5.(1) Insight + Saran/Action**

Berdasarkan Pie chart di atas menunjukkan bahwa:
- Customer yang tidak churn sebesar **73.46%** artinya potensi pelanggan untuk bertahan dalam menggunakan produk atau layanan perusahaan sebesar **73.46%**.
- Customer yang churn sebesar **26.54%** artinya potensi pelanggan untuk tidak bertahan dalam menggunakan produk atau layanan perusahaan sebesar **26.54%**
- Meskipun pelanggan churn rendah dibandingkan tidak churn, perusahaan tetap mengevaluasi dan melakukan perbaikan guna **meminimalisir potensi churn** dan **meningkatkan ketahanan pelanggan** dalam suatu produk atau layanan perusahaan.

Perlu melakukan visualisasi untuk Customer Churn reason, untuk melihat kenapa Churn bisa terjadi:

````python
fig = plt.subplots(figsize=(10,6))

df_reason = df['Churn Reason'].value_counts().reset_index()
df_reason.columns = ['Churn Reason', 'count']

ax = sns.barplot(data=df_reason, x='Churn Reason', y='count')
plt.title("Alasan Customer Melakukan Churn")
plt.xlabel('Alasan Churn')
plt.ylabel('Jumlah')

#teks pada x label
plt.xticks(rotation=45, horizontalalignment='right')

#Annonate value
for p in ax.patches:
  ax.annotate(text=format(p.get_height(), '.0f'),
                   xy=(p.get_x() + p.get_width() / 2, p.get_height()),
                   ha='center', va='center',
                   xytext=(0, 5),
                   textcoords='offset points',
                   color='black')
for bar in ax.patches:
  if bar.get_x() + bar.get_width()/2 in range(0,5): bar.set_facecolor('purple')

plt.show()
````

Pada Customer yang churn kebanyakan untuk 5 tertinggi adalah **attitude of support person, competitor offered higher download speeds, competitor offered more data, tidak tau, competitor made better offer.**. Solusi/Action adalah:

### **3.   Visualisasi untuk melihat 5 `kota` teratas yang customernya melakukan churn dan tidak melakukan churn**

````python
dff = df.groupby(['City', 'Churn Label']).size().reset_index(name='Count')
data_tinggi_churn = dff[dff['Churn Label'] == 'Yes'].nlargest(5, 'Count')
data_tinggi_nochurn = dff[dff['Churn Label'] == 'No'].nlargest(5, 'Count')
````

````python
dff.head()
````

#### Gambar 1

````python
#bar chart
# Membuat subplots
fig, axs = plt.subplots(1, 2, figsize=(20, 7))  # 1 baris, 2 kolom (0 , 1), 3 kolom (0, 1, 2)

#grafik 1
axs[0].bar(data_tinggi_churn['City'], data_tinggi_churn['Count'], label='Churn', color='red')

#teks untuk value
for x, y in zip(data_tinggi_churn['City'], data_tinggi_churn['Count']):
    axs[0].annotate(y, xy=(x, y), textcoords="offset points", xytext=(0, 5), ha='center', color='black')

axs[0].set_xlabel('Kota')
axs[0].set_ylabel('Frekuensi')
axs[0].set_title('Top 5 Tertinggi Customer Melakukan Churn')

#grafik 2
axs[1].bar(data_tinggi_nochurn['City'], data_tinggi_nochurn['Count'], label='No Churn', color='blue')

#teks untuk value
for x, y in zip(data_tinggi_nochurn['City'], data_tinggi_nochurn['Count']):
    axs[1].annotate(y, xy=(x, y), textcoords="offset points", xytext=(0, 5), ha='center', color='black')

axs[1].set_xlabel('Kota')
axs[1].set_ylabel('Frekuensi')
axs[1].set_title('Top 5 Tertinggi Customer yang Tidak Melakukan Churn')
plt.show()
````

#### Gambar 2

````python
# Menggabungkan data
merged_data = (pd.DataFrame(data_tinggi_churn) #data tinggi churn
               [['City', 'Count']]
               .rename(columns={'Count': 'Churn_Count'})
               .merge(pd.DataFrame(data_tinggi_nochurn) # data tinggi no churn
                      [['City', 'Count']]
                      .rename(columns={'Count': 'NoChurn_Count'}),
                      on='City',
                      how='outer')
               .fillna(0))

# Indeks untuk x-axis
x = np.arange(len(merged_data['City']))

# Lebar batang
width = 0.35

# Membuat grafik batang
fig, ax = plt.subplots(figsize=(12, 7))

# Bar untuk Churn
bar1 = ax.bar(x - width/2, merged_data['Churn_Count'], width, label='Churn', color='red')
# Bar untuk No Churn
bar2 = ax.bar(x + width/2, merged_data['NoChurn_Count'], width, label='No Churn', color='blue')

# Teks
for b in bar1:
    ax.annotate(f'{int(b.get_height())}',
                xy=(b.get_x() + b.get_width() / 2, b.get_height()),
                xytext=(0, 5),
                textcoords='offset points',
                ha='center',
                color='black')

for b in bar2:
    ax.annotate(f'{int(b.get_height())}',
                xy=(b.get_x() + b.get_width() / 2, b.get_height()),
                xytext=(0, 5),
                textcoords='offset points',
                ha='center',
                color='black')

# Label dan Judul
ax.set_xlabel('Kota')
ax.set_ylabel('Frekuensi')
ax.set_title('Top 5 Kota Tertinggi Pada Setiap Kategori Churn dan No Churn')
ax.set_xticks(x)
ax.set_xticklabels(merged_data['City'], rotation=45, ha='right')
ax.legend()

# Menampilkan grafik
plt.tight_layout()
plt.show()
````

#### **5. (3) Insight + Saran/Action**

- Secara keseluruhan, **Kategori Churn** untuk 5 kota tertinggi yang terpilih, yaitu **Los Angeles, Sacramento, San Diego, San Fransisco, San Jose, dan Sacramento**.

````python
g = sns.FacetGrid(df, col="Churn Label", height=6, aspect=1.5)
g.map_dataframe(sns.histplot, x="Tenure Months", kde=True)
g.set_axis_labels("Tenure Months")
g.set_ylabels("Jumlah")
g.set_titles("Churn Label: {col_name}")
# Judul Utama
g.figure.suptitle("Distribusi Tenure Months Berdasarkan Churn Label", fontsize=11, fontweight='bold')

# Jarak Judul Utama dengan Sub Judul
g.figure.subplots_adjust(top=0.85)
````

#### **5. (4) Insight + Saran/Action**

Berdasarkan Histogram dalam persebaran data di atas menunjukkan bahwa:
- Untuk **Tenure Months pada Churn Label = Yes** atau dalam artian Jumlah bulan yang telah dihabiskan customer di perusahaan pada akhir kuartal dengan tipe churn terlihat bahwa datannya persebarannya membentuk **Positive Skewness**. Artinya, nilai Modusnya lebih tinggi daripada Median daripada Mean (Modus>Median>Mean). **Nilai Modusnya pada nilai +- 750 an**. Hal ini, cenderung data tersebut mengindikasikan adanya **outlier/nilai ekstrem**. **Dalam konteksnya**, Customer melakukan churn untuk awal-awal periode begitu banyak jumlah pelanggannya dibandingkan periode lainnya (seiring menurun ke periode-periode selanjutnya kecenderungannya)
- Untuk **Tenure Months pada Churn Label = No** atau dalam artian Jumlah bulan yang telah dihabiskan customer di perusahaan pada akhir kuartal dengan tipe tidak churn terlihat bahwa datannya persebarannya membentuk karakteristik Bimodal**. Artinya, terdapat dua nilai Modusnya yang tinggi yaitu dengan jumlah kisaran **+- 620 an dan +- 850 an**. Dalam bimodal tersebut cenderung mengindikasikan tidak ada outlier karena **data-data antara dua puncaknya tidak saling jauh.**. **Dalam konteksnya**, Customer tidak melakukan churn pada awal bulan cukup banyak jumlah pelanggannya, namun kalau kita melihat dari periode-periode selanjutnya tidak begitu curam penurunannya dan ada mengalami kenaikan ketika di akhir-akhir periode, hingga menjadi jumlah pelanggan tertinggi pada periodenya ke-70 dibandingkan periode lainnya.

Mari kita cek outlier menggunakan Boxplot.

````python
# Boxplot persebaran tenure Months
sns.boxplot(data=df, x='Churn Label', y='Tenure Months', palette='muted')
plt.title("Boxplot Tenure Months Berdasarkan Label Churn")
plt.xlabel("Churn Label")
plt.ylabel("Tenure Months")
plt.show()
````

Terlihat bahwa setelah dicek Boxplot, untuk yang **Label Churn = Yes**, terdapat outlier jadi cocok dengan visualisasi Histogram sebelumnya. Sedangkan, untuk **Label Churn = No** tidak terdapat outlier.
- Lebar boxplotnya masih lebih tinggi **Label Churn = No**, karena sangat **beragam datanya** (datanya menyebar). Namun, untuk keduanya juga cenderung datanya sangat beragam (datanya menyebar) karena **lebar boxplot yang cukup lebar**. Hal ini mengindikasikan, **varians yang cukup besar**.

Saran/Action Perusahaan adalah:
- Perlu diketahui bahwa, pada periode awal-awal baik customer churn maupun no churn dimana jumlah pelanggannya cukup banyak (secara umum). Oleh karena itu, perusahaan harus meningkatkan kualitas baik dari segi produk ataupun pelayanan sebagai implementasi strategi perusahaan pada suatu produk untuk mempertahankan customer.
